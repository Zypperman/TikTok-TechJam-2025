````python
import cv2
import matplotlib.pyplot as plt

def view_image(path: str, figsize=(8, 8)) -> None:
    """Load and display an image *inline* in a Kaggle notebook."""
    img = cv2.imread(path)               # BGR
    if img is None:
        raise FileNotFoundError(f"Could not load image: {path}")

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=figsize)
    plt.axis('off')
    plt.imshow(img_rgb)
    plt.show()

# view_image(test_img_path)
view_image(test_rest_path)
````
